1 - Two Sum


```python
# 1. hash table 
class Solution:
    def twoSum(self, nums, target):
        if not nums:
            return [-1, -1]
        hash_dict = {}
        for i, num in enumerate(nums):
            if target - num in hash_dict:
                return [hash_dict[target - num], i]
            else: 
                hash_dict[num] = i
        return [-1, -1]
            
```

```python
# 2. use sort and two pointers 
class Solution:
    def twoSum(self, nums, target):
        if not nums:
            return [-1, -1]
        orig = nums[:]
        nums.sort()
        left, right = 0, len(nums) - 1 
        while left < right:
            if nums[left] + nums[right] > target:
                right -= 1 
            elif nums[left] + nums[right] < target:
                left += 1 
            else: 
                if nums[left] == nums[right]:
                    return sorted([loc for loc in range(len(orig)) if orig[loc] == nums[left]])
                else:
                    return [orig.index(nums[left]), orig.index(nums[right])]
        return [-1, -1]
        
```

2 - Add Two Numbers

You are given two non-empty linked lists representing two non-negative integers. The digits are stored in reverse order and each of their nodes contain a single digit. Add the two numbers and return it as a linked list.

You may assume the two numbers do not contain any leading zero, except the number 0 itself.

Example:

Input: (2 -> 4 -> 3) + (5 -> 6 -> 4)
    
Output: 7 -> 0 -> 8
    
Explanation: 342 + 465 = 807.

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def addTwoNumbers(self, l1, l2):
        """
        :type l1: ListNode
        :type l2: ListNode
        :rtype: ListNode
        """
        # corner case 
        if l1 is None or l2 is None:
            return ListNode(None)
        # dummy node 
        dummy = ListNode(-1)
        cur_node = dummy
        # if the sum of two values > 10, extras = 1 
        extras = 0
        # add
        l1_node, l2_node = l1, l2
        while l1_node is not None or l2_node is not None:
            l1_val = l1_node.val if l1_node is not None else 0 
            l2_val = l2_node.val if l2_node is not None else 0 
            
            sums = extras + l1_val + l2_val 
            
            extras = sums // 10 
            # update cur_node 
            cur_node.next = ListNode(sums % 10)
            cur_node = cur_node.next 
            # update l1 and l2 node 
            if l1_node is not None:
                l1_node = l1_node.next 
            if l2_node is not None:
                l2_node = l2_node.next 
            # if both list nodes reach the end and extras > 0     
            if l1_node is None and l2_node is None and extras > 0:
                cur_node.next = ListNode(extras)
        return dummy.next 
```

3 - Longest Substring Without Repeating Characters

Given a string, find the length of the longest substring without repeating characters.

Example 1:

Input: "abcabcbb"

Output: 3 

Explanation: The answer is "abc", with the length of 3. 

```python
# two pointers 
class Solution(object):
    def lengthOfLongestSubstring(self, s):
        """
        :type s: str
        :rtype: int
        """
        if not s:
            return 0 
        
        max_len = 0 
        
        for loc1 in range(len(s)):
            loc2 = loc1
            visited = set()
            while loc2 < len(s) and s[loc2] not in visited:
                visited.add(s[loc2])
                loc2 += 1 
            max_len = max(max_len, loc2 - loc1)
        return max_len 
                    
```

4 - Median of Two Sorted Arrays

There are two sorted arrays nums1 and nums2 of size m and n respectively.

Find the median of the two sorted arrays. The overall run time complexity should be O(log (m+n)).

You may assume nums1 and nums2 cannot be both empty.

Example 1:

nums1 = [1, 3]

nums2 = [2]

The median is 2.0

```python
# 1. a special case of find k-th value of two sorted arrays 
# depending on the total length of (nums1 + nums2), convert to find k-th value of two sorted arrays 
# drop the first k // 2 -th value from the array with smaller k // 2 -th value and keep search 
# the k - k // 2 -th value of the rest arrays, narrow down the search until find the first
# smaller element of two arrays 
import os 
class Solution(object):
    def findMedianSortedArrays(self, nums1, nums2):
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :rtype: float
        """
        # n > 0 as two arrays cannot be empty at the same time 
        n = len(nums1) + len(nums2)
        
        if n % 2 == 0:
            return float( self.find_kth(nums1, 0, nums2, 0, n // 2) + self.find_kth(nums1, 0, nums2, 0, n // 2 + 1) ) / 2 
        else:
            return self.find_kth(nums1, 0, nums2, 0, n // 2 + 1)
        
    def find_kth(self, nums1, start_1, nums2, start_2, k):
        # corner case , if the start index is bigger than the len of arrays 
        if start_1 >= len(nums1):
            # the k-th number is indexed by k - 1 
            return nums2[start_2 + k - 1]
        if start_2 >= len(nums2):
            return nums1[start_1 + k - 1]
        
        # recursion termination condition  
        if k == 1:
            return min(nums1[start_1], nums2[start_2])
        
        # check the k // 2 -th value of each array counted from the start index 
        # the array with smaller half k value must not contain median (k-th value) as two arrays are sorted 
        half_k_1 = nums1[start_1 + k // 2 - 1] if start_1 + k // 2 - 1 < len(nums1) else sys.maxsize
        half_k_2 = nums2[start_2 + k // 2 - 1] if start_2 + k // 2 - 1 < len(nums2) else sys.maxsize
        
        # compare and drop the first k // 2 elements for the array with smaller half k value 
        if half_k_1 < half_k_2:
            return self.find_kth(nums1, start_1 + k // 2, nums2, start_2, k - k // 2)
        else:
            return self.find_kth(nums1, start_1, nums2, start_2 + k // 2, k - k // 2)
```

```python
# 2. binary search on answer 
# first take the min and max of two arrays and get a median value of the range (min + max) / 2 
# second count how many number are smaller/bigger then this median value 
# keep narrowing down the range of median value until find it 
class Solution(object):
    def findMedianSortedArrays(self, nums1, nums2):
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :rtype: float
        """
        # n > 0 as two arrays cannot be empty at the same time 
        n = len(nums1) + len(nums2)
        if n % 2 == 0:
            return float( self.find_kth(nums1, nums2, n // 2) + self.find_kth(nums1, nums2, n // 2 + 1) ) / 2 
        else:
            return self.find_kth(nums1, nums2, n // 2 + 1)
    
    def find_kth(self, nums1, nums2, k):    
        # corner case 
        if len(nums1) == 0:
            return nums2[k - 1]
        if len(nums2) == 0:
            return nums1[k - 1]
        
        # lower, upper bounds of values in the two arrays, not index 
        lower, upper = min(min(nums1), min(nums2)), max(max(nums1), max(nums2))
        while lower + 1 < upper:
            # array have only integers 
            mid = (lower + upper) // 2 
            
            # number of values <= current mid 
            count1 = sum([1  if nums1[i] <= mid else 0 for i in range(len(nums1))])
            count2 = sum([1 if nums2[i] <= mid else 0 for i in range(len(nums2))])
            
            if count1 + count2 < k:
                lower = mid 
            else:
                upper = mid 
            
        # now lower + 1 >= upper 
        count1 = sum([1  if nums1[i] <= lower else 0 for i in range(len(nums1))])
        count2 = sum([1  if nums2[i] <= lower else 0 for i in range(len(nums2))])
        
        # if count of values <= lower is >= k, then lower is the k-th number 
        # the > case exists when the lower number happens to appear multiple times 
        # e.g., [1, 1], [1, 2], count(x <= 1) = 3 with k = 2 
        if count1 + count2 >= k:
            print(lower)
            return lower
        else:
            print(upper)
            return upper
```

5 - Longest Palindromic Substring

Given a string s, find the longest palindromic substring in s. You may assume that the maximum length of s is 1000.



```python
# 1. use two pointers, start from each character and spread to two directions 
class Solution(object):
    def longestPalindrome(self, s):
        """
        :type s: str
        :rtype: str
        """
        # corner case 
        if s is None or len(s) == 0:
            return ''
        
        longest = ''
        # loop 
        for mid in range(len(s)):
            # find odd and even length palindrome with middle point at s[mid]
            for i in range(2):
                # print('mid + i', mid + i)
                substr = self.find_palindrome(s, mid, mid + i)
                if len(substr) > len(longest):
                    longest = substr 
        return longest
    
    # find palindrome that start from s[left], s[right] and spread to the two opposite 
    # left and right directions of s 
    def find_palindrome(self, s, left, right):
        while left >= 0 and right < len(s) and s[left] == s[right]:
            left -= 1 
            right += 1
        
        return s[left + 1: right]
```

```python
# 2. dynamic programming 
# dp[i][j]: boolean value that represent that if string[i : j+1] is palindrome or not
# dp[i][i] = True as a character itself is always palindrome 
# first fill dp[0][1], dp[1][2], ..., dp[i][i + 1], then fill dp[0][2], dp[1, 3], ..., dp[i][i + 2], ...
# dp[i][j] = True if dp[i + 1][j - 1] is True and s[i] = s[j] and corresponding length of palindrome if j - i + 1

class Solution(object):
    def longestPalindrome(self, s):
        """
        :type s: str
        :rtype: str
        """
        # corner case 
        if s is None or len(s) == 0:
            return ''
        # initialilzation
        # the worst case is the longest palindrome contains only one charaters
        longest = s[0]
        n = len(s)
        dp = [[False for _ in range(n)] for _ in range(n)]
        for i in range(n):
            dp[i][i] = True
            # set dp[i][i - 1] = True as the dp[start + 1][end - 1] will be used when palindrom_len = 1
            dp[i][i - 1] = True
        
    
        # loop over all possible palindrome lengths 
        for palindrome_len in range(1, n):
            # loop for start as any possible location of the palindrome 
            for start in range(n - palindrome_len):
                end = start + palindrome_len 
                dp[start][end] = dp[start + 1][end - 1] and s[start] == s[end]
                # check if the current substr is palindrome and update longest palindrome if qualified
                if dp[start][end] and end - start + 1 > len(longest):
                    longest = s[start : end + 1]
        return longest
        
```

6 - ZigZag Conversion

The string "PAYPALISHIRING" is written in a zigzag pattern on a given number of rows like this: (you may want to display this pattern in a fixed font for better legibility)



```python
# Find the pattern of how the zigzag characters are arranged:
# 1. Think 
# A   E
# B D F H
# C   G   
# Treat A, B , C and D as a group, row indices are 0, 1, 2, 2
# 2. The total number chars of this group is 2 * numRows - 2 as it takes two columns, but there are 
#    no chars in top and bottom rows 
# 3. row_idx = i % (2 * numRows - 2) if row_idx < numRow (e.g., i = 0, 1, 2 --> row_idx = 0, 1, 2)
#    row_idx = 2 * numRows - 2 - row_idx(from above) (e.g., the above row_idx = 3 for s[3], but 
#        numRows = 3 with max row_idx = 2, so row_idx for s[3] is 4 - 3 = 1)

class Solution(object):
    def convert(self, s, numRows):
        """
        :type s: str
        :type numRows: int
        :rtype: str
        """
        # cornercase 
        if numRows <= 1 or numRows >= len(s):
            return s 
        # intialize zigzag characters by rows 
        zigzag = ['' for _ in range(numRows)]
        
        n = len(s)
        # find row_idx of each char in s 
        for i in range(n):
            row_idx = i % (2 * numRows - 2)
            # update zigzag
            if row_idx < numRows:
                zigzag[row_idx] += s[i]
            else:
                zigzag[2 * numRows - 2 - row_idx] += s[i]
        # print(zigzag)
        # merge them together 
        results = ''
        for string in zigzag:
            results += string
        return results 
```

7 - Reverse Integer

Given a 32-bit signed integer, reverse digits of an integer.

Assume we are dealing with an environment which could only store integers within the 32-bit signed integer range: $$[−2^{31},  2^{31} − 1]$$ 

```python
class Solution(object):
    def reverse(self, x):
        """
        :type x: int
        :rtype: int
        """
        if x is None:
            return None

        # find if x is negative 
        negative_ind = True if x < 0 else False 
        
        # set x to be non-negative 
        if x < 0:
            x = -x 
            
        digits = []
        while x > 0:
            digits.append(x % 10)
            x //= 10 
        new = 0 
        for i in range(len(digits)):
            new = new * 10 + digits[i]
        # add sign 
        new = -new if negative_ind else new 
        
        # check if new is out of range 
        if new < -2**31 or new > 2**31 - 1:
            return 0 
        return new 
```

9 - Palindrome Number

Determine whether an integer is a palindrome. An integer is a palindrome when it reads the same backward as forward.

Example 1:

Input: 121
    
Output: true

```python
class Solution(object):
    def isPalindrome(self, x):
        """
        :type x: int
        :rtype: bool
        """
        # corner case 
        if x is None or x < 0:
            return False 
        
        # if x has only one digit 
        if x // 10 == 0:
            return True

        # get each digit of number x 
        s = []        
        while x > 0:
            s.append(x % 10)
            x //= 10
        # two pointers 
        left, right = 0, len(s) - 1 
        while left < right:
            if s[left] != s[right]:
                return False 
            else:
                left += 1 
                right -= 1
        return True
```

10 -  Regular Expression Matching

Given an input string (s) and a pattern (p), implement regular expression matching with support for '.' and '*'.

'.' Matches any single character.

'*' Matches zero or more of the preceding element.

The matching should cover the entire input string (not partial).

Note:

s could be empty and contains only lowercase letters a-z.

p could be empty and contains only lowercase letters a-z, and characters like . or *.

Example 4:

Input:
s = "aab"

p = "c\*a\*b"

Output: true

Explanation: c can be repeated 0 times, a can be repeated 1 time. Therefore it matches "aab".

```python
# 1. search with memoization 
class Solution(object):
    def isMatch(self, s, p):
        """
        :type s: str
        :type p: str
        :rtype: bool
        """
        return self.is_match_helper(s, 0, p, 0, {})
    
    def is_match_helper(self, s, i, p, j, memo):
        # memo 
        if (i, j) in memo:
            return memo[(i, j)]
        
        if i == len(s):
            # test if pattern p[j: ] can be matched with an empty input string
            # e.g, input is "", pattern is "c*c*"
            return self.match_empty(p[j: ])
        
        if j == len(p):
            return False
        # j can jump to j + 2 directly so j may > len(p)
        if j > len(p):
            return
        
        if j + 1 < len(p) and p[j + 1] == '*':
            # current s[i] and p[j] match, and s[i + 1:] and p[j] match (* can still match zero or more new chars )
            # current s[i] and p[j] not match, but s[i:] and p[j + 2:] match 
            # (* repeats the previous element 0 time)
            # e.g., s = 'aab', p = 'a*b' or p = 'aa*b'
            matched = (self.is_match_char(s[i], p[j]) and self.is_match_helper(s, i + 1, p, j, memo)) or self.is_match_helper(s, i, p, j + 2, memo)
        else:
            matched = self.is_match_char(s[i], p[j]) and self.is_match_helper(s, i + 1, p, j + 1, memo)
        # update memo     
        memo[(i, j)] = matched
        
        return matched 
    
    def is_match_char(self, s, p):
        return s == p or p == '.'
    
    # only check location 1, 3, 5, ...
    def match_empty(self, pattern):
        # if there are only odd num of chars, it cannot be * (we tell if p[j + 1] is * or not, p[j] is always not *),
        # return False
        if len(pattern) % 2 == 1:
            return False 
        for i in range(len(pattern) // 2):
            if pattern[i * 2 + 1] != '*':
                return False 
        return True
        
```

```python
# 2. dynamic memoization
# dp[i + 1][j + 1] = dp[i][j] if s[i] == p[j] or p[j] == '.'
# if p[j] == '*'
#  - if s[i] = p[j - 1] or p[j - 1] == '.', dp[i + 1][j + 1] = dp[i][j + 1] (* repeats p[j-1] one time)
#  - or dp[i + 1][j + 1] = dp[i + 1][j - 1] (* maps p[j - 1] zero times)
class Solution(object):
    def isMatch(self, s, p):
        """
        :type s: str
        :type p: str
        :rtype: bool
        """
        n, m = len(s), len(p)
        
        # initialization 
        dp = [[False for _ in range(m + 1)] for _ in range(n + 1)]
        dp[0][0] = True    
        
        # a*, a*b*, a*b*c* 
        for i in range(2, len(p) + 1):
            if p[i - 1] == "*":
                dp[0][i] = dp[0][i - 2]
        
        # loop to update 
        for i in range(1, n + 1):
            for j in range(1, m + 1):
                # s[i - 1] and p[j - 1] match
                if s[i - 1] == p[j - 1] or p[j - 1] == '.':
                    dp[i][j] = dp[i - 1][j - 1]
                # if s[i - 1] and p[j - 1] doesn't match, but p[j - 1] is *
                if p[j - 1] == '*':
                    # * matches zero char 
                    dp[i][j] = dp[i][j - 2]
                    # start matches one or more chars 
                    if s[i - 1] == p[j - 2] or p[j - 2] == '.':
                        dp[i][j] |= dp[i - 1][j]
        return dp[n][m]
```

11 - Container With Most Water

Given n non-negative integers a1, a2, ..., an , where each represents a point at coordinate (i, ai). n vertical lines are drawn such that the two endpoints of line i is at (i, ai) and (i, 0). Find two lines, which together with x-axis forms a container, such that the container contains the most water.

Input: [1,8,6,2,5,4,8,3,7]
    
Output: 49

```python
# Two pointers 
class Solution(object):
    def maxArea(self, height):
        """
        :type height: List[int]
        :rtype: int
        """
        if not height:
            return 0 
        
        left, right = 0, len(height) - 1 
        max_area  = 0 
        while left < right:
            cur_area = min(height[left], height[right]) * (right - left)
            # move pointers 
            # if height[left] is smaller, than we should keep height[right] for 
            # looking for max_area 
            if cur_area > max_area:
                max_area = cur_area
            if height[left] > height[right]:
                right -= 1 
            else:
                left += 1 
        return max_area
        
```

12 - Integer to Roman

Roman numerals are represented by seven different symbols: I, V, X, L, C, D and M.

```python
# https://www.youtube.com/watch?v=pSxbXcCH5no
class Solution(object):
    def intToRoman(self, num):
        """
        :type num: int
        :rtype: str
        """
        if num is None:
            return 
        # possible values that can be represented in roman numbers
        values = [1000, 900, 500, 400, 100, 90, 50, 40, 10, 9, 5, 4, 1]
        # corresponding roman numbers 
        roman = ['M', 'CM', 'D', 'CD', 'C', 'XC', 'L', 'XL', 'X', 'IX', 'V', 'IV', 'I']
        
        results = []
        for i in range(len(values)):
            while num >= values[i]:
                num -= values[i]
                results.append(roman[i])
        return ''.join(results)
```

13 - Roman to Integer


```python
# set a map with roman number as key and integer as values and loop through the roman number
# used one pattern of roman numbers:
# for two characters, if the left one is smaller than the right one
# then the corresponding integer is right - left 
class Solution(object):
    def romanToInt(self, s):
        """
        :type s: str
        :rtype: int
        """
        if s is None or len(s) == 0:
            return 0 
        hash_map = {'M':1000, 'CM':900, 'D':500,
                    'CD':400, 'C':100, 'XC':90, 
                    'L':50, 'XL':40, 'X':10, 'IX':9,
                    'V':5, 'IV':4, 'I':1}
        # roman = ['M', 'CM', 'D', 'CD', 'C', 'XC', 'L', 'XL', 'X', 'IX', 'V', 'IV', 'I']
        # values = [1000, 900, 500, 400, 100, 90, 50, 40, 10, 9, 5, 4, 1]
        
        num = 0 
        # loop 
        i = 0
        while i < len(s):
            # i + 1 is out of range, the last of the loop 
            if i == len(s) - 1:
                num += hash_map[s[i]]
                break
            # check if the next roman number is bigger than the current one 
            if hash_map[s[i + 1]] > hash_map[s[i]]:
                num += hash_map[s[i :i+2]]
                i += 2 
            else:
                num += hash_map[s[i]]
                i += 1 
        return num 
```

14 - Longest Common Prefix

Write a function to find the longest common prefix string amongst an array of strings.

If there is no common prefix, return an empty string "".

Example 1:

Input: ["flower","flow","flight"]
    
Output: "fl"

```python
class Solution(object):
    def longestCommonPrefix(self, strs):
        """
        :type strs: List[str]
        :rtype: str
        """
        # corner case 
        if len(strs) == 0:
            return ''
        if len(strs) == 1:
            return strs[0]
        # >= 2 strings 
        prefix = ''
        for i in range(len(strs[0])):
            char = strs[0][i]
            for j in range(1, len(strs)):
                if i > len(strs[j]) - 1 or strs[j][i] != char:
                    return prefix
            # if every string has the same i-th char, update prefix 
            prefix += char 
        return prefix 
```

15 - 3Sum

Given an array nums of n integers, are there elements a, b, c in nums such that a + b + c = 0? Find all unique triplets in the array which gives the sum of zero.

Example:

Given array nums = [-1, 0, 1, 2, -1, -4],

A solution set is:
[
  [-1, 0, 1],
  [-1, -1, 2]

```python
# 1. loop through on number, convert to two sum problem 
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        if not nums:
            return []
        # sort 
        nums.sort()
        
        results = []
        
        # loop 
        for i in range(len(nums) - 2):
            # remove duplicates 
            if i > 0 and nums[i] == nums[i - 1]:
                continue 
            # two sum problem with target == -nums[i]
            two_sums = self.two_sum(nums[i + 1 : ], -nums[i])
            # if there are elements of two_sums, add them to results  
            if two_sums:
                for item in two_sums:
                    results.append([nums[i]] + list(item))
        return results
    
    def two_sum(self, nums, target):
        # hash table to save visited num 
        hash_dict = {}
        # save unique two sum pairs 
        res = set()
        for i, num in enumerate(nums):
            if target - num in hash_dict:
                res.add((target - num, num))
            else:
                hash_dict[num] = i 
        return res 
 
```

```python
# Record prev.next, prev.next.next, prev.next.next.next as a, b ,c  and set a = b, b = a, a = c
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def swapPairs(self, head: ListNode) -> ListNode:
        if head is None:
            return head 
        # create a dummy node 
        dummy = ListNode(-1)
        dummy.next = head 
        
        # initialize prev node 
        prev = dummy 
        while prev.next and prev.next.next:
            a = prev.next 
            b = a.next 
            c = b.next 
            
            a.next = c 
            b.next = a 
            prev.next = b 
            
            # update prev for next pair's swap 
            prev = a 
        return dummy.next 
```

34 - Find First and Last Position of Element in Sorted Array

Given an array of integers nums sorted in ascending order, find the starting and ending position of a given target value.

Your algorithm's runtime complexity must be in the order of O(log n).

If the target is not found in the array, return [-1, -1].



```python
class Solution(object):
    def searchRange(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        if not nums or target is None:
            return [-1, -1]
        
        
        
        # binary search 
        left, right = 0, len(nums) - 1
        # find the first position first 
        while left + 1 < right:
            mid = (left + right) // 2 
            if nums[mid] < target:
                left = mid 
            else:
                right = mid 
        
        if nums[left] == target:
            first = left 
        elif nums[right] == target:
            first = right 
        else:
            return [-1, -1]
        # find the last position 
        last = first 
        while last + 1  < len(nums) and nums[last + 1] == target:
            last += 1 
        return [first, last]
```

The count-and-say sequence is the sequence of integers with the first five terms as following:

1.     1
2.     11

5.     111221

```python
# use recursion
class Solution(object):
    def countAndSay(self, n):
        """
        :type n: int
        :rtype: str
        """
        if not n:
            return 
        if n == 1:
            return '1'
        # n > 1, recursion, obtain how it looks like for the case of n - 1  
        n_minus_1 = self.countAndSay(n - 1)
        
        # find the count of each consecutive repeated number and find the results for the case of n
        results = []
        i = 0
        while i < len(n_minus_1):
            char = n_minus_1[i]
            count = 0 
            cur_char = char
            while i < len(n_minus_1) and cur_char == n_minus_1[i]:
                count += 1 
                i += 1 
            results.append(str(count))
            results.append(char)
        return ''.join(results)
```

```python
# dfs on combinations 
class Solution(object):
    def combinationSum(self, candidates, target):
        """
        :type candidates: List[int]
        :type target: int
        :rtype: List[List[int]]
        """
        if candidates is None or target is None:
            return []
        
        # 
        results = []
        self.dfs(candidates, 0, target, [], results)
        return results 
    
    def dfs(self, candidates, start_index, target, path, results):
        # append
        if sum(path) == target:
            results.append(path[:])
        # stop if the sum of path values is > target 
        if sum(path) > target:
            return 
        
        # loop
        for i in range(start_index, len(candidates)):
            path.append(candidates[i])
            self.dfs(candidates, i, target, path, results)
            path.pop()
        
```

42 - Trapping Rain Water

Given n non-negative integers representing an elevation map where the width of each bar is 1, compute how much water it is able to trap after raining.

```python
# Two pointers 
# check the water that can be trapped at each location separately 
# to compute the water that can be trapped at location i, 
# from left:  max(height[0:i]) - height[i] 
# from right: max(height[i+1 : ]) - height[i]
class Solution(object):
    def trap(self, height):
        """
        :type height: List[int]
        :rtype: int
        """
        if not height:
            return 0 
        
        left, right = 0, len(height) - 1 
        left_max, right_max = 0, 0 
        water = 0 
        while left  < right:
            if height[left] < height[right]:
                left_max = max(left_max, height[left])
                water += left_max - height[left]
                left += 1 
            else:
                right_max = max(right_max, height[right])
                water += right_max - height[right]
                right -= 1 
            
        return water 
```

44 - Wildcard Matching

Given an input string (s) and a pattern (p), implement wildcard pattern matching with support for '?' and '*'.

'?' Matches any single character.

'*' Matches any sequence of characters (including the empty sequence).

The matching should cover the entire input string (not partial).

Note:

s could be empty and contains only lowercase letters a-z.

p could be empty and contains only lowercase letters a-z, and characters like ? or *.

* DFS, dynamic programming

```python
# 1.  memoization search
# check if s[:i + 1] and p[: j + 1] match
class Solution(object):
    def isMatch(self, s, p):
        """
        :type s: str
        :type p: str
        :rtype: bool
        """
        return self.is_match_helper(s, 0, p, 0, {})
    # check if s[ : i + 1] and p[ : j + 1] matches
    def is_match_helper(self, s, i, p, j, memo):
        # check memo 
        if (i, j) in memo:
            return memo[(i, j)]
        
        # if the index i out of range for s, that is, all chars in s have been checked 
        if len(s) == i:
            # if len(p) != j, only when p[j: ] contain only * (match 0 char), it's true
            for index in range(j, len(p)):
                if p[index] != '*':
                    return False
            # if all p[j: ] are all * 
            return True 
        
        # if len(p) == j, pattern has nothing left, while s is not finished 
        if len(p) == j:
            return False
        
        # now both i and i are within range
        # discuss two cases 
        if p[j] != '*':
            matched = self.is_match_char(s[i], p[j]) and self.is_match_helper(s, i + 1, p, j + 1, memo)
        else:
            # * matches zero or one char
            matched = self.is_match_helper(s, i + 1, p, j, memo) or self.is_match_helper(s, i, p, j + 1, memo)
        # add memo 
        memo[(i, j)] = matched
        
        return matched
    
    # is matched when the two chars are the same or pattern is ? (can match any character)
    def is_match_char(self, s, p):
        return s == p or p == '?'
        
```

```python
# 2. dynamic programming
# dp[i][j] = dp[i - 1][j - 1] if s[i] == p[j] or p[j] == '?'
# dp[i][j] = dp[i][j - 1] or dp[i - 1][j] if p[j] == '*'
class Solution(object):
    def isMatch(self, s, p):
        """
        :type s: str
        :type p: str
        :rtype: bool
        """
        n = len(s)
        m = len(p)
        # initialization 
        dp = [[False for _ in range(m + 1)] for _ in range(n + 1)]
        dp[0][0] = True
        
        # corner case: if source is empty and patter contains only * 
        if n == 0 and p.count('*') == m:
            return True
        
        # n != 0, loop to update dp table 
        for i in range(n + 1):
            for j in range(m + 1):
                if i > 0 and j > 0:
                    dp[i][j] |= dp[i - 1][j - 1] and (s[i - 1] == p[j - 1] or p[j - 1] == '?')
                
                if i > 0 and j > 0:
                    dp[i][j] |= dp[i - 1][j] and p[j - 1] == '*'
                    
               # i = 0, j > 0, dp[i][j] is initialized as False for any i, j except i = 0, j= 0 
                if j > 0:
                    # if p[j - 1] == '*', can match zero or more than chars 
                    dp[i][j] |= dp[i][j - 1] and p[j - 1] == '*'
        return dp[n][m]
        
```

45 - Jump Game II


Given an array of non-negative integers, you are initially positioned at the first index of the array.

Each element in the array represents your maximum jump length at that position.

Example:

Input: [2,3,1,1,4]

Output: 2

Explanation: The minimum number of jumps to reach the last index is 2.
    Jump 1 step from index 0 to 1, then 3 steps to the last index.

```python
# greedy 
class Solution(object):
    def jump(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        # edge cases 
        if not nums:
            return 0 
        # currrent end can be reached with one jump at current position 
        cur_end = 0 
        # furthest loc can be reached with one more jump from any position that can be reached 
        # from current jump 
        furthest_loc = 0 
        # min number of jumps 
        jump = 0
        # loop through all locations BEFORE the last 
        for i in range(len(nums) - 1):
            if i + nums[i] > furthest_loc:
                furthest_loc = i + nums[i]
            if i == cur_end:
                jump += 1 
                cur_end = furthest_loc 
        return jump
            
        
```

48 - Rotate Image

You are given an n x n 2D matrix representing an image.

Rotate the image by 90 degrees (clockwise).

```python
# 1. exchange matrix[i][j] with matrix[j][i]
# 2. reverse each row of the matrix
class Solution(object):
    def rotate(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: None Do not return anything, modify matrix in-place instead.
        """
        if matrix is None or len(matrix) == 0 or len(matrix[0]) == 0 :
            return []
        
        for i in range(0, len(matrix)):
            for j in range(i + 1, len(matrix[0])):
                matrix[i][j], matrix[j][i] = matrix[j][i], matrix[i][j]
        
        # reverse 
        for row in matrix:
            row.reverse()
            
        return matrix
```

49 - Group Anagrams

Given an array of strings, group anagrams together.

Example:

Input: ["eat", "tea", "tan", "ate", "nat", "bat"],

Output:

[
  ["ate","eat","tea"],
  
  ["nat","tan"],
  
  ["bat"]
]

```python
# Use a fixed (e.g., sorted) version of strings as hash table key and 
# the corresponding anagram strs as values list, so to group anagram strs together
class Solution(object):
    def groupAnagrams(self, strs):
        """
        :type strs: List[str]
        :rtype: List[List[str]]
        """
        if not strs:
            return []
        
        # 
        hash_table = {}
        for word in strs:
            key = ''.join(sorted(word))
            hash_table.setdefault(key, [])
            hash_table[key].append(word)
        return hash_table.values()
        
```
