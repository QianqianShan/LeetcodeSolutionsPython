953 - Verifying an Alien Dictionary

In an alien language, surprisingly they also use english lowercase letters, but possibly in a different order. The order of the alphabet is some permutation of lowercase letters.

Given a sequence of words written in the alien language, and the order of the alphabet, return true if and only if the given words are sorted lexicographicaly in this alien language.

 

Example 1:

Input: words = ["hello","leetcode"], order = "hlabcdefgijkmnopqrstuvwxyz"
    
Output: true
    
Explanation: As 'h' comes before 'l' in this language, then the sequence is sorted.

```python
class Solution(object):
    def isAlienSorted(self, words, order):
        """
        :type words: List[str]
        :type order: str
        :rtype: bool
        """
        if not words:
            return True 
        # add hash_table with character as key, its order as index 
        hash_table = {}
        for i, char in enumerate(order):
            hash_table[char] = i
        # compare each pair of words by its order in words 
        for i in range(len(words) - 1):
            w1 = words[i]
            w2 = words[i + 1]
            # compare 
            common_len = min(len(w1), len(w2))
            for j in range(common_len):
                # if the characters of two words at location j are the same
                # compare the next location 
                if w1[j] == w2[j]:
                    continue
                # if w1[j] index is bigger than w2[j] index 
                if hash_table[w1[j]] > hash_table[w2[j]]:
                    return False 
                # if w1[j] index is smaller than w2[j] index 
                # the current pair order is correct 
                else:
                    break
            # if all common_len character are the same for w1 and w2,
            # if len(w1) is bigger then len(w2), False
            if w1[ :common_len] == w2 and len(w1) > len(w2):
                return False 
        return True
    
```

965 -  Univalued Binary Tree

A binary tree is univalued if every node in the tree has the same value.

Return true if and only if the given tree is univalued.



```python
# 1. Recursion 
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def isUnivalTree(self, root):
        """
        :type root: TreeNode
        :rtype: bool
        """
        if not root:
            return True
        uni_left = (root.left is None or root.val == root.left.val) and self.isUnivalTree(root.left)
        uni_right = (root.right is None or root.val == root.right.val) and self.isUnivalTree(root.right)
        return uni_left and uni_right 
```

```python
# 2. DFS: collect all node values and check if they are the same 
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def isUnivalTree(self, root):
        """
        :type root: TreeNode
        :rtype: bool
        """
        if not root:
            return True
        results = []
        self.dfs(root, results)
        # print(results)
        return len(set(results)) == 1
    
    def dfs(self, node, results):
        if node is None:
            return 
        # node is not None, append values 
        results.append(node.val)
        self.dfs(node.left, results)
        self.dfs(node.right, results)
        
        
        
            
```

973 - K Closest Points to Origin

We have a list of points on the plane.  Find the K closest points to the origin (0, 0).

(Here, the distance between two points on a plane is the Euclidean distance.)

You may return the answer in any order.  The answer is guaranteed to be unique (except for the order that it is in.)

 

Example 1:

Input: points = [[1,3],[-2,2]], K = 1
    
Output: [[-2,2]]

```python
# maintain a heap of size K 
# push the negative distance into a heap so whenever there is a point that has 
# smallest negative distance will be poped out if size of heap > K  
from heapq import heappush, heappop 
class Solution(object):
    def kClosest(self, points, K):
        """
        :type points: List[List[int]]
        :type K: int
        :rtype: List[List[int]]
        """
        if not points:
            return []
        # push points info to heap 
        heap = []
        for point in points:
            dist = point[0]**2 + point[1]**2 
            # push -distance^2, -x, -y 
            # -x, -y is used to break ties when distances are the same 
            heappush(heap, (-dist, -point[0], -point[1]))
            # drop points that are no longer possible to be top K closest 
            if len(heap) > K:
                heappop(heap)
        # read point data from heap 
        results = []
        while heap:
            cur_point = heappop(heap)
            results.append([-cur_point[1], -cur_point[2]])
        return results 
            
```

977 - Squares of a Sorted Array

Given an array of integers A sorted in non-decreasing order, return an array of the squares of each number, also in sorted non-decreasing order.

 

Example 1:

Input: [-4,-1,0,3,10]
    
Output: [0,1,9,16,100]

```python
# two pointers
# 1. split the sorted array into two parts: negative and positive 
# 2. add squares to new list one by one like merge sort 
class Solution(object):
    def sortedSquares(self, A):
        """
        :type A: List[int]
        :rtype: List[int]
        """
        if A is None:
            return 
        slow, fast = 0, 0 
        # fast pointer points to the first position >= 0 
        # slow points to the last position < 0 
        while fast < len(A) and A[fast] < 0:
            fast += 1 
        slow = fast - 1 
        # print('slow', slow)
        # print(fast)
        results = []
        while slow >= 0 and fast < len(A):
            if A[slow]**2 > A[fast]**2:
                results.append(A[fast]**2)
                fast += 1 
            else:
                results.append(A[slow]**2)
                slow -= 1 
        # in case slow doesn't reach 0 
        while slow >= 0:
            results.append(A[slow]**2)
            slow -= 1 
        while fast < len(A):
            results.append(A[fast]**2)
            fast += 1 
        return results 
                
    
```

993 - Cousins in Binary Tree


In a binary tree, the root node is at depth 0, and children of each depth k node are at depth k+1.

Two nodes of a binary tree are cousins if they have the same depth, but have different parents.

We are given the root of a binary tree with unique values, and the values x and y of two different nodes in the tree.

Return true if and only if the nodes corresponding to the values x and y are cousins.

 

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None
from collections import deque 
class Solution(object):
    def isCousins(self, root, x, y):
        """
        :type root: TreeNode
        :type x: int
        :type y: int
        :rtype: bool
        """
        if not root:
            return 
        
        # bfs, record node value and its parents 
        queue = deque([(root, None)])
        while queue:
            size = len(queue)
            # reset x_parent, y_parent values for each level to make sure that 
            # x and y are in the same level 
            x_parent, y_parent = None, None
            # loop values in the same level 
            for _ in range(size):
                cur_node, parent = queue.popleft()
                # record x_parent/y_parent if available 
                if cur_node.val == x:
                    x_parent = parent 
                    
                if cur_node.val == y:
                    y_parent = parent 
                # update queue with next level's nodes 
                if cur_node.left:
                    queue.append((cur_node.left, cur_node.val))
                if cur_node.right:
                    queue.append((cur_node.right, cur_node.val))
            # print('x', x_parent)
            # print('y', y_parent)
            if x_parent is not None and y_parent is not None and x_parent != y_parent:
                return True
        return False 
            
```

994 - Rotting Oranges

In a given grid, each cell can have one of three values:

the value 0 representing an empty cell;

the value 1 representing a fresh orange;

the value 2 representing a rotten orange.

Every minute, any fresh orange that is adjacent (4-directionally) to a rotten orange becomes rotten.

Return the minimum number of minutes that must elapse until no cell has a fresh orange.  If this is impossible, return -1 instead.



Input: [[2,1,1],[1,1,0],[0,1,1]]
    
Output: 4

```python
from collections import deque 
directions = [[0, 1], [0, -1], [-1, 0], [1, 0]]
class Solution(object):
    def orangesRotting(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        # corner case, if not grid, or all elements in grid is 0 
        if not grid or sum([sum(row) for row in grid]) == 0:
            return 0 
        # compute the minutes 
        # initialization 
        queue = deque([])
        visited = set()
        # find all current 2s 
        for i in range(len(grid)):
            for j in range(len(grid[0])):
                if grid[i][j] == 2:
                    queue.append((i, j))
                    visited.add((i, j))
        minute = 0 
        while queue:
            # print(queue)
            size = len(queue)
            for _ in range(size):
                cur_x, cur_y = queue.popleft()
                # infect neighbors 
                for delta_x, delta_y in directions:
                    new_x, new_y = cur_x + delta_x, cur_y + delta_y 
                    if (new_x, new_y) in visited:
                        continue 
                    if self.is_valid(grid, new_x, new_y):
                        grid[new_x][new_y] = 2 
                        queue.append((new_x, new_y))
                        visited.add((new_x, new_y))
            minute += 1 
                    
            
        # scan grid to see if there are still fresh orange (1) left 
        for i in range(len(grid)):
            for j in range(len(grid[0])):
                if grid[i][j] == 1:
                    return -1 
        # for the last step of queue, all oranges have been infected and cannot find further neighbors 
        # to infect, so should -1 
        return minute - 1 
    
    
    def is_valid(self, grid, new_x, new_y):
        return 0 <= new_x < len(grid) and 0 <= new_y < len(grid[0]) and grid[new_x][new_y] == 1
        
```

997 - Find the Town Judge

In a town, there are N people labelled from 1 to N.  There is a rumor that one of these people is secretly the town judge.

If the town judge exists, then:

The town judge trusts nobody.

Everybody (except for the town judge) trusts the town judge.

There is exactly one person that satisfies properties 1 and 2.

You are given trust, an array of pairs trust[i] = [a, b] representing that the person labelled a trusts the person labelled b.

If the town judge exists and can be identified, return the label of the town judge.  Otherwise, return -1.

 

Example 1:

Input: N = 2, trust = [[1,2]]
    
Output: 2

```python
# graph, assume x is Judge
# trust[i] = [a --> x],
# then judge should satisfy two properties 
# 1. x has (N- 1) indegrees 
# 2. no trust[i] has [x --> ?] as x trust nobody 
class Solution(object):
    def findJudge(self, N, trust):
        """
        :type N: int
        :type trust: List[List[int]]
        :rtype: int
        """
        if not trust:
            if N == 1:
                return N 
            else: 
                return -1 
        indegrees = {}
        not_judge = set()
        for item in trust:
            indegrees.setdefault(item[1], 0)
            indegrees[item[1]] += 1 
            not_judge.add(item[0])
        for person in indegrees:
            if indegrees[person] == N - 1 and person not in not_judge:
                return person 
        return -1 
        
```
