1108 - Defanging an IP Address

Given a valid (IPv4) IP address, return a defanged version of that IP address.

A defanged IP address replaces every period "." with "[.]".

 

Example 1:

Input: address = "1.1.1.1"
    
Output: "1[.]1[.]1[.]1"

```python
# create a new string 
class Solution(object):
    def defangIPaddr(self, address):
        """
        :type address: str
        :rtype: str
        """
        if not address:
            return ''
        results = ''
        for char in address:
            if char == '.':
                results += '[.]'
            else:
                results += char 
        return results 
        
```

1114 - Print in Order

Suppose we have a class:

public class Foo {
  public void first() { print("first"); }
  public void second() { print("second"); }
  public void third() { print("third"); }
}

The same instance of Foo will be passed to three different threads. Thread A will call first(), thread B will call second(), and thread C will call third(). Design a mechanism and modify the program to ensure that second() is executed after first(), and third() is executed after second().

 

Example 1:

Input: [1,2,3]

Output: "firstsecondthird"

Explanation: There are three threads being fired asynchronously. The input [1,2,3] means thread A calls first(), thread B calls second(), and thread C calls third(). "firstsecondthird" is the correct output.

```python
class Foo(object):
    def __init__(self):
        self.first_print = False
        self.second_print = False
        pass


    def first(self, printFirst):
        """
        :type printFirst: method
        :rtype: void
        """
        
        # printFirst() outputs "first". Do not change or remove this line.
        printFirst()
        self.first_print = True


    def second(self, printSecond):
        """
        :type printSecond: method
        :rtype: void
        """
        while not self.first_print:
            continue 
        
        # printSecond() outputs "second". Do not change or remove this line.
        printSecond()
        self.second_print = True
            
            
    def third(self, printThird):
        """
        :type printThird: method
        :rtype: void
        """
        while not self.second_print:
            continue 
        # printThird() outputs "third". Do not change or remove this line.
        printThird()
        
    
```

1119 - Remove Vowels from a String

Given a string S, remove the vowels 'a', 'e', 'i', 'o', and 'u' from it, and return the new string.

 

Example 1:

Input: "leetcodeisacommunityforcoders"

Output: "ltcdscmmntyfrcdrs"


```python
class Solution(object):
    def removeVowels(self, S):
        """
        :type S: str
        :rtype: str
        """
        if not S:
            return ""
        vowels = set(['a', 'e', 'i', 'o', 'u'])
        new_str = ""
        for char in S:
            if char in vowels:
                continue 
            new_str += char
        return new_str
```

1122 - Relative Sort Array

Given two arrays arr1 and arr2, the elements of arr2 are distinct, and all elements in arr2 are also in arr1.

Sort the elements of arr1 such that the relative ordering of items in arr1 are the same as in arr2.  Elements that don't appear in arr2 should be placed at the end of arr1 in ascending order.

 

Example 1:

Input: arr1 = [2,3,1,3,2,4,6,7,9,2,19], arr2 = [2,1,4,3,9,6]
    
Output: [2,2,2,1,4,3,3,9,6,7,19]

```python
# hash table to count the appearance of each number
class Solution(object):
    def relativeSortArray(self, arr1, arr2):
        """
        :type arr1: List[int]
        :type arr2: List[int]
        :rtype: List[int]
        """
        if not arr1 or not arr2:
            return []
        hash_table = {}
        for num in arr1:
            if num not in hash_table:
                hash_table[num] = 1
            else:
                hash_table[num] += 1 
        results = []
        # first add elements with the order of nums in arr2 
        for num in arr2:
            results += [num for _ in range(hash_table[num])]
            del hash_table[num]
        # add elements not in arr2 
        for num in sorted(hash_table):
            results += [num for _ in range(hash_table[num])]
        return results 
        
```

1128 - Number of Equivalent Domino Pairs


Given a list of dominoes, dominoes[i] = [a, b] is equivalent to dominoes[j] = [c, d] if and only if either (a==c and b==d), or (a==d and b==c) - that is, one domino can be rotated to be equal to another domino.

Return the number of pairs (i, j) for which 0 <= i < j < dominoes.length, and dominoes[i] is equivalent to dominoes[j].

 

Example 1:

Input: dominoes = [[1,2],[2,1],[3,4],[5,6]]

Output: 1

```python
class Solution(object):
    def numEquivDominoPairs(self, dominoes):
        """
        :type dominoes: List[List[int]]
        :rtype: int
        """
        if not dominoes:
            return 0
        hash_table = {}
        for nums in dominoes: 
            nums.sort()
            key =  (nums[0], nums[1])
            if key in hash_table:
                hash_table[key] += 1 
            else:
                hash_table[key] = 1 
        # print(hash_table)
        count = 0 
        for key in hash_table:
            if hash_table[key] > 1:
                count += (hash_table[key] * (hash_table[key] - 1))/2
                # print(count)
        return count 
```

1133 - Largest Unique Number

Given an array of integers A, return the largest integer that only occurs once.

If no integer occurs once, return -1.

 

Example 1:

Input: [5,7,3,9,4,9,8,3,1]
    
Output: 8

```python
class Solution(object):
    def largestUniqueNumber(self, A):
        """
        :type A: List[int]
        :rtype: int
        """
        if not A:
            return -1 
        hash_table = {}
        for num in A:
            if num not in hash_table:
                hash_table[num] = 1 
            else:
                hash_table[num] += 1 
        uniq_nums = []
        
        for item in hash_table:
            if hash_table[item] == 1:
                uniq_nums.append(item)
        return -1 if not uniq_nums else max(uniq_nums)
        
```
