451 - Sort Characters By Frequency

Given a string, sort it in decreasing order based on the frequency of characters.

Example 1:

Input:
"tree"

Output:
"eert"

Explanation:
    
'e' appears twice while 'r' and 't' both appear once.
So 'e' must appear before both 'r' and 't'. Therefore "eetr" is also a valid answer.

```python
from heapq import heappush, heappop 
class Solution(object):
    def frequencySort(self, s):
        """
        :type s: str
        :rtype: str
        """
        if not s:
            return ''
        hash_table = {}
        for char in s:
            hash_table.setdefault(char, 0)
            hash_table[char] += 1 
        heap = []
        for char in hash_table:
            heappush(heap, (-hash_table[char], char))
        results = []
        while heap:
            freq, char = heappop(heap)
            results.append(char * (-freq))
        return ''.join(results) 
        
```

454 - 4Sum II

Given four lists A, B, C, D of integer values, compute how many tuples (i, j, k, l) there are such that A[i] + B[j] + C[k] + D[l] is zero.

```python
class Solution:
    def fourSumCount(self, A: List[int], B: List[int], C: List[int], D: List[int]) -> int:
        # corner case 
        if A is None or B is None or C is None or D is None:
            return  0
        # hash_table, key = sum(A[i] + B[j]), value = counts of key appearances 
        hash_table = {}
        for i in range(len(A)):
            for j in range(len(B)):
                hash_table.setdefault(A[i] + B[j], 0)
                hash_table[A[i] + B[j]] += 1 
        # loop over C and D 
        count = 0 
        for i in range(len(C)):
            for j in range(len(D)):
                new_target = - (C[i] + D[j])
                if new_target in hash_table:
                    count += hash_table[new_target] 
        return count 
```

463 - Island Perimeter

You are given a map in form of a two-dimensional integer grid where 1 represents land and 0 represents water.

Grid cells are connected horizontally/vertically (not diagonally). The grid is completely surrounded by water, and there is exactly one island (i.e., one or more connected land cells).

The island doesn't have "lakes" (water inside that isn't connected to the water around the island). One cell is a square with side length 1. The grid is rectangular, width and height don't exceed 100. Determine the perimeter of the island.

 

Example:

Input:
    
[[0,1,0,0],
 
 [1,1,1,0],
 
 [0,1,0,0],
 
 [1,1,0,0]]

Output: 16

```python
# count the total number of edges all 1s and minus number of neighbor 1s for each 1 
class Solution(object):
    def islandPerimeter(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        if not grid or not grid[0]:
            return 0 
        ones = 0 
        neighbors = 0 
        
        row, col = len(grid), len(grid[0])
        for i in range(row):
            for j in range(col):
                if grid[i][j] == 1:
                    ones += 1 
                    if i > 0 and grid[i - 1][j] == 1:
                        neighbors += 1 
                    if i < row - 1 and grid[i + 1][j] == 1:
                        neighbors += 1 
                    if j > 0 and grid[i][j - 1] == 1:
                        neighbors += 1 
                    if j < col - 1 and grid[i][j + 1] == 1:
                        neighbors += 1 
        return ones * 4 - neighbors 
```

496 - Next Greater Element I

You are given two arrays (without duplicates) nums1 and nums2 where nums1â€™s elements are subset of nums2. Find all the next greater numbers for nums1's elements in the corresponding places of nums2.

The Next Greater Number of a number x in nums1 is the first greater number to its right in nums2. If it does not exist, output -1 for this number.

Example 1:
    
Input: nums1 = [4,1,2], nums2 = [1,3,4,2].
    
Output: [-1,3,-1]

```python
# 1. brute force 
class Solution(object):
    def nextGreaterElement(self, nums1, nums2):
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :rtype: List[int]
        """
        if not nums1:
            return []
        results = []
        for num in nums1:
            loc = nums2.index(num)
            first_greater = -1 
            for i in range(loc, len(nums2)):
                if nums2[i] > num:
                    first_greater = nums2[i]
                    break
            results.append(first_greater)
        return results 
```

```python
# 2. use stack and hashmap 
# hashmap key = values in nums2, value = the first number bigger than key on its right, if not exists, -1 
# use stack to store the current num when looping through nums2 

class Solution(object):
    def nextGreaterElement(self, nums1, nums2):
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :rtype: List[int]
        """
        if not nums1:
            return []
        stack = []
        hash_map = {}
        for num in nums2:
            while stack and num > stack[-1]:
                hash_map[stack.pop()] = num
            # push each num into stack 
            stack.append(num)
        # if stack is not empty, the existing elements in stack don't have larger values on their right 
        for num in stack:
            hash_map[num] = -1 
        
        # find key-value pairs for num in nums1 
        results = []
        for num in nums1:
            results.append(hash_map[num])
        return results 
```
