51 - N-Queens

The n-queens puzzle is the problem of placing n queens on an n×n chessboard such that no two queens attack each other. Given an integer n, return all distinct solutions to the n-queens puzzle.





```python
# use three sets to save the appeared used col, row + col and row - col 
class Solution:
    def solveNQueens(self, n: int) -> List[List[str]]:
        if not n:
            return []
        
        self.results = []
        self.search(n, 0, set(), set(), set(), [])
        return self.results 
    
    def search(self, n, row, cols, diffs, sums, path):
        # when row is out of range (no elements can be added)
        if row == n:
            self.results.append(self.draw_board(path))
            return 
        
        for col in range(n):
            if col in cols or (row - col) in diffs or (row + col) in sums:
                continue 
            cols.add(col)
            diffs.add(row - col)
            sums.add(row + col)
            path.append(col)
            self.search(n, row + 1, cols, diffs, sums, path)
            path.pop()
            sums.remove(row + col)
            diffs.remove(row - col)
            cols.remove(col)
    # convert col nums to the required shape 
    def draw_board(self, path):
        res = []
        for col in path:
            res.append(''.join(['.' if i != col else 'Q' for i in range(len(path))]))
        return res 
            
```

53 - Maximum Subarray

Given an integer array nums, find the contiguous subarray (containing at least one number) which has the largest sum and return its sum.

Example:

Input: [-2,1,-3,4,-1,2,1,-5,4],

Output: 6

Explanation: [4,-1,2,1] has the largest sum = 6.

```python
# prefix sum 
import sys 
class Solution(object):
    def maxSubArray(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        if nums is None:
            return 
        
        max_sum = - sys.maxsize 
        min_sum = 0 
        prefix_sum = 0
        
        for i in range(len(nums)):
            prefix_sum += nums[i]
            max_sum = max(max_sum, prefix_sum - min_sum)
            min_sum = min(min_sum, prefix_sum)
        return max_sum
```

```python
# first take the spiral circle when we can (start_row < end_row and start_col < end_col)
# three cases after the spiral circles:
# 1. there is only one number left (start_row == end_row and start_col == end_col)
# 2. there is one row of numbers left (start_row == end_row and start_col < end_col)
# 3. there is one col of numbers left (start_row < end_row and start_col == end_col)
class Solution(object):
    def spiralOrder(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: List[int]
        """
        # corner case 
        if matrix is None or len(matrix) == 0 or len(matrix[0]) == 0:
            return []
        
        #
        start_row, end_row = 0, len(matrix) - 1
        start_col, end_col = 0, len(matrix[0]) - 1
        
        results = []
        while start_row < end_row and start_col < end_col:
            # print('start', start_row)
            # print('end', end_row)
            for j in range(start_col, end_col):
                results.append(matrix[start_row][j])

            for i in range(start_row, end_row):
                results.append(matrix[i][end_col])

            for k in range(end_col, start_col, -1):
                results.append(matrix[end_row][k])
            # print(results)
            for l in range(end_row, start_row, -1):
                results.append(matrix[l][start_col])
            start_row += 1 
            start_col += 1 
            end_row -= 1 
            end_col -= 1
        # print(results)
        # print('start', start_row)
        # print('end', end_row)
        # print('start col', start_col)
        # print('end col', end_col)
        # 
        # discuss three cases after the spiral circles 
        if start_row == end_row:
            for i in range(start_col, end_col + 1):
                results.append(matrix[start_row][i])
        elif start_col == end_col:
            for i in range(start_row, end_row + 1):
                results.append(matrix[i][start_col])
        

        return results
    
```

55 - Jump Game

Given an array of non-negative integers, you are initially positioned at the first index of the array.

Each element in the array represents your maximum jump length at that position.

Determine if you are able to reach the last index.

Example 1:

Input: [2,3,1,1,4]
    
Output: true

```python
class Solution(object):
    def canJump(self, nums):
        """
        :type nums: List[int]
        :rtype: bool
        """
        if not nums:
            return False 

        max_idx = 0
        i = 0 
        while i < len(nums) - 1: 
            # furthest idx can be reached 
            max_idx = max(i + nums[i], max_idx)
            # if i can be reached given the current max_idx, check the next i and see if you could update max_idx
            if max_idx > i:
                i += 1 
            else:
                return False
        return True 
```

56 - Merge Intervals


Given a collection of intervals, merge all overlapping intervals.

Example 1:

Input: [[1,3],[2,6],[8,10],[15,18]]
    
Output: [[1,6],[8,10],[15,18]]
    
Explanation: Since intervals [1,3] and [2,6] overlaps, merge them into [1,6].

```python
# first sort all intervals by their start values so we could focus on the end values when comparing 
# compare a new interval's end value with the end value of the last interval added into results and 
# add intervals to results accordingly 
class Solution(object):
    def merge(self, intervals):
        """
        :type intervals: List[List[int]]
        :rtype: List[List[int]]
        """
        
        # corner case 
        if not intervals:
            return []
        # sort intervals by start values 
        intervals.sort(key = lambda x:x[0])
        # intervals = sorted(intervals, key = lambda x:x[0])
        results = []
        #
        for interval in intervals:
            # if there is no intervals in results or 
            # the current START interval value is bigger than the end value of last interval
            # (no overlap of current interval and the last interval in results)
            if len(results) ==  0 or interval[0] > results[-1][1]:
                results.append(interval)
            # if there is overlap  
            else:
                results[-1][1] = max(results[-1][1], interval[1])
        return results
```

60 - Permutation Sequence

The set [1,2,3,...,n] contains a total of n! unique permutations.

By listing and labeling all of the permutations in order, we get the following sequence for n = 3:

"123"

"132"

"213"

"231"

"312"

"321"

Given n and k, return the kth permutation sequence.

Note:

Given n will be between 1 and 9 inclusive.

Given k will be between 1 and n! inclusive.


```python
# 1. Find all permutations and sort, find the kth 
# NOTE: the difference of str(i) and i in visited and path 
class Solution(object):
    def getPermutation(self, n, k):
        """
        :type n: int
        :type k: int
        :rtype: str
        """
        if not n or not k:
            return ''
        results = []
        self.dfs(n, [], set(), results)
        results.sort()
        return results[k - 1]
    
    def dfs(self, n, path, visited, results):
        if len(path) == n:
            results.append(''.join(path))
            return 
        for i in range(1, n + 1):
            if i in visited:
                continue
            path.append(str(i))
            visited.add(i)
            self.dfs(n, path, visited, results)
            visited.remove(i)
            path.pop()
            
```

```python
# 2.  cantor expansion https://zh.wikipedia.org/wiki/%E5%BA%B7%E6%89%98%E5%B1%95%E5%BC%80
class Solution(object):
    def getPermutation(self, n, k):
        """
        :type n: int
        :type k: int
        :rtype: str
        """
        if not n or not k:
            return ''
        # factorials of 0!, 1!, 2!, ..., n!
        factorial = [1]
        for i in range(1, n + 1):
            factorial.append(factorial[-1] * i)
        permutation = []
        all_elements = list(range(1, n + 1))
        for i in range(n):
            digit = (k - 1) / factorial[n - i -1]
            permutation.append(all_elements[digit])
            all_elements.remove(all_elements[digit])
            k = (k - 1) % factorial[n - i - 1] + 1 
        return ''.join([str(x) for x in permutation])
            
 
```

61 - Rotate List

Given a linked list, rotate the list to the right by k places, where k is non-negative.

Example 1:

Input: 1->2->3->4->5->NULL, k = 2
    
Output: 4->5->1->2->3->NULL
    
Explanation:
    
rotate 1 steps to the right: 5->1->2->3->4->NULL
    
rotate 2 steps to the right: 4->5->1->2->3->NULL

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def rotateRight(self, head, k):
        """
        :type head: ListNode
        :type k: int
        :rtype: ListNode
        """
        if not head or not k:
            return head 
        
        # count the total length of linked list, n 
        n = 0
        start = head 
        while start is not None:
            n += 1 
            start = start.next 
        if n == 1:
            return head 
        # take k = k % n 
        k = k % n 
        if k == 0:
            return head 
        
        # rotate
        start = head 
        count = 0 
        prev = None
        while count < n - k:
            count += 1 
            prev = start
            start = start.next 
        new_head = start
        # print(new_head.val)
        if prev is not None:
            prev.next = None 
        
        # concatenate new head and old head 
        new_start = new_head 
        while new_start is not None and new_start.next is not None:
            new_start = new_start.next 
        new_start.next = head 
        
        return new_head 
        
```

62 - Unique Paths

A robot is located at the top-left corner of a m x n grid (marked 'Start' in the diagram below).

The robot can only move either down or right at any point in time. The robot is trying to reach the bottom-right corner of the grid (marked 'Finish' in the diagram below).

How many possible unique paths are there?

From pure math: there are a total of (n - 1) + (m - 1) steps needed from start to end, choose (n - 1) from (n -1 + m - 1) to go right. 

```python
# dynamic programming 
# dp[i][j] = # of paths from (0, 0) to (i, j)
# dp[i][j] = dp[i - 1][j] + dp[i][j - 1] --- (0, 0) --> (i, j) must pass (i-1, j) or (i, j - 1)
# return dp[m - 1][n - 1]
class Solution(object):
    def uniquePaths(self, m, n):
        """
        :type m: int
        :type n: int
        :rtype: int
        """
        if not m or not n:
            return 
        # initialization 
        dp = [[0 for _ in range(n)] for _ in range(m)]
        # there is only one path from (0, 0) --> (0, i)
        for i in range(n):
            dp[0][i] = 1
        # similarly 
        for i in range(m):
            dp[i][0] = 1 
        
        for i in range(1, m):
            for j in range(1, n):
                dp[i][j] = dp[i - 1][j] + dp[i][j - 1]
        return dp[m-1][n-1]
```

64 - Minimum Path Sum

Given a m x n grid filled with non-negative numbers, find a path from top left to bottom right which minimizes the sum of all numbers along its path.

Note: You can only move either down or right at any point in time.

Example:

Input:
    
[
  [1,3,1],
    
  [1,5,1],
    
  [4,2,1]
]

Output: 7
    
Explanation: Because the path 1→3→1→1→1 minimizes the sum.


```python
# 1. dfs, find all path sum and return the min  
class Solution(object):
    def minPathSum(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        if not grid:
            return 0 
        nrow = len(grid)
        ncol = len(grid[0])
        results = []
        self.dfs(grid, 0, 0, nrow - 1, ncol - 1, [grid[0][0]], results)
        return min(results)
    
    def dfs(self, grid, start_x, start_y, end_x, end_y, path, results):
        if start_x == end_x and start_y == end_y:
            results.append(sum(path))
            return 
        
        for delta_x, delta_y in [[0, 1], [1, 0]]:
            new_x, new_y = start_x + delta_x, start_y + delta_y
            if not self.is_valid(grid, new_x, new_y):
                continue 
            path.append(grid[new_x][new_y])
            self.dfs(grid, new_x, new_y, end_x, end_y, path, results)
            path.pop()
    
    def is_valid(self, grid, x, y):
        return 0 <= x < len(grid) and 0 <= y < len(grid[0])
        
```

```python
# 2. dp 
# dp[i][j]: the min sum from (0, 0) to (i, j)
# dp[i][j] = min(dp[i - 1][j], dp[i][j - 1]) + grid[i][j]
# return dp[nrow - 1][ncol - 1]
class Solution(object):

    def minPathSum(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        if not grid:
            return 0 
        nrow = len(grid)
        ncol = len(grid[0])
        # 
        dp = [[None for _ in range(ncol)] for _ in range(nrow)]
        for i in range(nrow):
            for j in range(ncol):
                if i == 0 and j == 0:
                    dp[i][j] = grid[i][j]
                elif i == 0 and j > 0:
                    dp[i][j] = dp[i][j - 1] + grid[i][j]
                elif j == 0 and i > 0:
                    dp[i][j] = dp[i - 1][j] + grid[i][j]
                else:
                    dp[i][j] = min(dp[i - 1][j], dp[i][j - 1]) + grid[i][j]
        return dp[nrow - 1][ncol - 1]
```

66 -  Plus One

Given a non-empty array of digits representing a non-negative integer, plus one to the integer.

The digits are stored such that the most significant digit is at the head of the list, and each element in the array contain a single digit.

You may assume the integer does not contain any leading zero, except the number 0 itself.

Example 1:

Input: [1,2,3]
    
Output: [1,2,4]
    
Explanation: The array represents the integer 123.

```python
# special case: [9, 9, 9] + 1 ---> [1, 0, 0, 0]
class Solution(object):
    def plusOne(self, digits):
        """
        :type digits: List[int]
        :rtype: List[int]
        """
        if digits is None:
            return digits 
        n = len(digits)
        # if the last digit < 9, add 1 directly 
        if digits[n - 1] < 9:
            digits[n - 1] += 1 
            return digits 
        # if the last digit == 9, move left until finding the digit which is < 9 
        i = n - 1
        while i >= 0 and digits[i] == 9:
            digits[i] = 0
            i -= 1 
        # i < 0 (eg, [9,9,9]) add extra 1 on the left 
        if i < 0:
            digits = [1] + digits
        else:
            digits[i] += 1 
        return digits 

            
    
```

67 -  Add Binary

Given two binary strings, return their sum (also a binary string).

The input strings are both non-empty and contains only characters 1 or 0.

Example 1:

Input: a = "11", b = "1"
    
Output: "100"
    
Example 2:

Input: a = "1010", b = "1011"
    
Output: "10101"

See how binary adds up at https://www.youtube.com/watch?v=jB_sRh5yoZk


```python
# See how binary adds up at https://www.youtube.com/watch?v=jB_sRh5yoZk
# 1010 
#+1011
#=10101
class Solution(object):
    def addBinary(self, a, b):
        """
        :type a: str
        :type b: str
        :rtype: str
        """
        i, j = len(a) - 1, len(b) - 1 
        carry = 0 
        reverse_res = []
        while i >= 0 and j >= 0:
            int_a, int_b = int(a[i]), int(b[j])
            cur_digit = (int_a + int_b + carry) % 2 
            carry = (int_a + int_b + carry) // 2 
            reverse_res.append(str(cur_digit))
            i -= 1 
            j -= 1 
            # print(reverse_res)
        while i >= 0:
            int_a = int(a[i])
            cur_digit = (int_a + carry) % 2 
            carry = (int_a + carry) // 2 
            reverse_res.append(str(cur_digit))
            i -= 1 
        while j >= 0:
            int_b = int(b[j])
            cur_digit = (int_b + carry) % 2
            carry = (int_b + carry) // 2 
            reverse_res.append(str(cur_digit))
            j -= 1 
        if carry == 1:
            reverse_res.append(str(carry))
        return ''.join(reversed(reverse_res))
            
            
        
```

69 - Sqrt(x)

Implement int sqrt(int x).



```python
class Solution:
    def mySqrt(self, x: int) -> int:
        if x == 0:
            return 0 
        left, right = 0, x 
        while (right - left) >= 1e-4:
            mid = (left + right) / 2 
            if mid**2 > x:
                right = mid
            else:
                left = mid
        if int(right)**2 <= x:
            return int(right)
        else:
            return int(left)
```

76 - Minimum Window Substring

Given a string S and a string T, find the minimum window in S which will contain all the characters in T in complexity O(n).

Example:

Input: S = "ADOBECODEBANC", T = "ABC"
    
Output: "BANC"

```python

class Solution:
    # @return a string
    def minWindow(self, S, T):
        # set two hash maps
        # t_map: key: distinct chars in T; value: # of appearances 
        # s_map: key: distinct chars to be matched for T;  value: # of appearnces needed to be match 
        s_map = {}
        t_map = {}
        for char in T:
            t_map.setdefault(char, 0)
            t_map[char] += 1 
        # s_map is the same as t_map at the begginning 
        s_map = t_map.copy()
        
        # the total # of chars needs to be matched for T 
        match_cnt = len(T)
        # left pointer 
        start = 0
        # the position of the left char of matched substring in S 
        min_start = 0
        # set min_len of substring as a number bigginer than the length of S 
        # (an impossible length if there exists matched substring)
        min_len = len(S) + 1 
        
        # move the end pointer from left to right 
        for end in range(len(S)):
            # if the current char exists in T, matched, corresponding s_map minus 1 
            if S[end] in t_map and t_map[S[end]] > 0:
                s_map[S[end]] -= 1 
                # if T needs S[end], match_cnt -= 1
                #  when s_map[S[end]] < 0, the current S substing has already have enought S[end]
                if s_map[S[end]] >= 0:
                    match_cnt -= 1 
            # when every char in T is contained in current S substring 
            # find the current min_len (move start to its right until it cannot move)
            if match_cnt == 0:
                while True:
                    if S[start] in t_map and t_map[S[start]] > 0:
                        if s_map[S[start]] < 0:
                            s_map[S[start]] += 1 
                        else:
                            break
                    start += 1 
                # update min_len 
                if min_len > end - start + 1:
                    min_len = end - start + 1 
                    min_start = start 
        # if min_len is not updated            
        if min_len == len(S) + 1:
            return ''
        else:
            return S[min_start: min_start + min_len]
                
```

79 - Word Search

Given a 2D board and a word, find if the word exists in the grid.

The word can be constructed from letters of sequentially adjacent cell, where "adjacent" cells are those horizontally or vertically neighboring. The same letter cell may not be used more than once.

```python
# use a prefix set to mimic trie 
DIRECTIONS = [(0, 1), (0, -1), (-1, 0), (1, 0)]

class Solution:
    def exist(self, board: List[List[str]], word: str) -> bool:
        # corner case 
        if board is None or len(board) == 0 or len(board[0]) == 0:
            return False
        
        # prefixes of word (including itself) 
        prefix_set = set()
        for i in range(len(word)):
            prefix_set.add(word[ :i + 1])
        
        # loop through board 
        for i in range(len(board)):
            for j in range(len(board[0])):
                if self.search(board, i, j, board[i][j] ,word, prefix_set, set([(i, j)])):
                    return True 
        return False 
    
    def search(self, board, row, col, substr,word, prefix_set, visited):
        if substr == word:
            return True 
        for delta_x, delta_y in DIRECTIONS:
            new_row, new_col = row + delta_x, col + delta_y
            if not self.is_valid(board, new_row, new_col):
                continue 
            if (new_row, new_col) in visited:
                continue 
            new_str = substr + board[new_row][new_col]
            if new_str in prefix_set:
                visited.add((new_row, new_col))
                if self.search(board, new_row, new_col, new_str, word, prefix_set, visited):
                    return True 
                else:
                    # backtrack 
                    visited.remove((new_row, new_col))
    def is_valid(self, board, x, y):
        return 0 <= x < len(board) and 0 <= y < len(board[0])
```

Example 1:

```python
class Solution(object):
    def search(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: bool
        """
        if nums is None or len(nums) == 0 or target is None:
            return 
        left, right = 0, len(nums) - 1 
        while left + 1 < right:
            mid = (left + right) // 2 
            if nums[mid] < nums[right]:
                if nums[mid] <= target <= nums[right]:
                    left = mid 
                else:
                    right = mid 
            # duplicates
            elif nums[mid] == nums[right]:
                right -= 1 
            else:
                if nums[left] <= target <= nums[mid]:
                    right = mid 
                else:
                    left = mid 
        return nums[left] == target or nums[right] == target
```

Example 1:

Input: "12"
    
Output: 2
    
Explanation: It could be decoded as "AB" (1 2) or "L" (12).

```python
# 1. dynamic programming
# dp[n] the total number of ways to decode s with len(s) is n 
# dp[n] = dp[n - 1] + dp[n - 2] when s[n] is not zero and s[n - 1: (n + 1)] is in [10, 26]
# dp[0] = 1 
class Solution(object):
    def numDecodings(self, s):
        """
        :type s: str
        :rtype: int
        """
        if not s:
            return 0 
        # intialization 
        dp = [0] * (len(s) + 1)
        
        dp[0] = 1 
        
        # turn 'abc' to ['a', 'b', 'c']
        sc = list(s)
        
        for i in range(1, len(s) + 1):
            # check if the cur_char sc[i-1] is zero or not 
            # if it's zero, then s[i-1] itself is not decodable, must be together with its previous char
            if sc[i - 1] != '0':
                dp[i] += dp[i - 1]
            # when i - 2 >= 0, one can check if two chars sc[i - 2], sc[i - 2] are valid for decoding
            if i >= 2:
                two_chars_val = int(sc[i - 2]) * 10 + int(sc[i - 1])
                # valid only if the value is between [10, 26]
                if 10 <= two_chars_val <= 26:
                    dp[i] += dp[i - 2]
        return dp[len(s)]
```

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def inorderTraversal(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        if root is None:
            return 
        dummy = TreeNode(-1)
        dummy.right = root 
        stack = [dummy]
        results = []
        while stack:
            cur_node = stack.pop()
            if cur_node.right:
                cur_node = cur_node.right 
                while cur_node:
                    stack.append(cur_node)
                    cur_node = cur_node.left
            if stack:
                results.append(stack[-1].val)
        return results
            
```

98 - Validate Binary Search Tree

Given a binary tree, determine if it is a valid binary search tree (BST).

Assume a BST is defined as follows:

* The left subtree of a node contains only nodes with keys less than the node's key.
* The right subtree of a node contains only nodes with keys greater than the node's key.
* Both the left and right subtrees must also be binary search trees.
 

Example:

      2
     / \
    1   3

Input: [2,1,3]

Output: true

```python
# 1. compare the inorder traverse results with sorted values in the tree

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def isValidBST(self, root: TreeNode) -> bool:
        if not root:
            return True 
        inorder_vals = self.in_order_traverse(root)
        # use set() to make sure that there are no duplicates 
        return inorder_vals == sorted(list(set(inorder_vals)))
    def in_order_traverse(self, root):
        if root is None:
            return []
        return self.in_order_traverse(root.left) + [root.val] + self.in_order_traverse(root.right)
```

```python
# 2. inorder iterate over the tree and compare each node value with its previous node value  

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def isValidBST(self, root: TreeNode) -> bool:
        if not root:
            return True 
        self.prev = None 
        return self.helper(root)
    
    def helper(self, root):
        if root is None:
            return True 
        # if root is not None 
        
        # iterate over left subtree 
        if not self.helper(root.left):
            return False 
        
        # if left subtree is valid 
        if self.prev is not None and self.prev.val >= root.val:
            return False 
        
        # if both left subtree and root are valid, check right subtree
        
        # update prev node as root 
        self.prev = root 
        return self.helper(root.right)
```

```python
# 3. Recursion 
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def isValidBST(self, root: TreeNode) -> bool:
        if not root:
            return True 
        return self.valid_tree(root)[0]
    
    def valid_tree(self, root):
        if root is None:
            return True, None, None
        left_ind, left_min, left_max = self.valid_tree(root.left)
        right_ind, right_min, right_max = self.valid_tree(root.right)
        
        # case 1
        if not left_ind or not right_ind:
            return False, None, None 
        # case 2 
        if left_max is not None and root.val <= left_max:
            return False, None, None 
        # case 3
        if right_min is not None and root.val >= right_min:
            return False, None, None 
        
        # if all above three cases are note satisfied, then both left and right subtrees are valid 
        # and the current tree with root as root is also valid 
        min_val = left_min if left_min is not None else root.val 
        max_val = right_max if right_max is not None else root.val 
        return True, min_val, max_val
        
        if left_ind and right_ind and root.val > left_max and root.val < right_min:
            return True, left_min, right_max 
        return False, None, None  
        
```
