653 - Two Sum IV - Input is a BST

Given a Binary Search Tree and a target number, return true if there exist two elements in the BST such that their sum is equal to the given target.

Example 1:

Input: 

       5
      / \
     3   6
    / \   \
    2   4   7

Target = 9

Output: True

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def findTarget(self, root, k):
        """
        :type root: TreeNode
        :type k: int
        :rtype: bool
        """
        if not root or k is None:
            return 
        nums = self.in_order(root)
        left, right = 0, len(nums) - 1 
        while left < right:
            while left < right and nums[left] + nums[right] > k:
                right -= 1 
            while left < right and nums[left] + nums[right] < k:
                left += 1 
            if left < right and nums[left] + nums[right] == k:
                return True 
        return False 
    
    def in_order(self, node):
        if not node:
            return []
        return self.in_order(node.left) + [node.val] + self.in_order(node.right)
```

674 - Longest Continuous Increasing Subsequence

Given an unsorted array of integers, find the length of longest continuous increasing subsequence (subarray).

Example 1:

Input: [1,3,5,4,7]

Output: 3

Explanation: The longest continuous increasing subsequence is [1,3,5], its length is 3. 

Even though [1,3,5,7] is also an increasing subsequence, it's not a continuous one where 5 and 7 are separated by 4. 

```python
class Solution(object):
    def findLengthOfLCIS(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        if nums is None or len(nums) == 0:
            return 0
        longest = 1 
        cur_len = 1 
        for i in range(len(nums) - 1):
            if nums[i + 1] > nums[i]:
                cur_len += 1 
                longest = max(longest, cur_len)
            else:
                cur_len = 1 
        return longest 
        
```

680 - Valid Palindrome II

Given a non-empty string s, you may delete at most one character. Judge whether you can make it a palindrome.


Example:

Input: "abca"

Output: True

Explanation: You could delete the character 'c'.

```python
class Solution(object):
    def validPalindrome(self, s):
        """
        :type s: str
        :rtype: bool
        """
        # corner case 
        if s is None:
            return False 
        if len(s) == 0:
            return True 
        
        # loop 
        left, right = 0, len(s) - 1 
        while left < right:
            # if not equal 
            if s[left] != s[right]:
                # tell if the unvisited substrings can be a palindrome by removing one character 
                return self.is_palindrome(s[left : right]) or self.is_palindrome(s[left + 1 : right + 1])
            else:
                # two pointers goes closer until finding unequal characters 
                left += 1 
                right -= 1 
        # if every character is symmetric 
        return True 
    
    def is_palindrome(self, s):
        if len(s) == 0:
            return True 
        left, right = 0, len(s) - 1 
        while left < right:
            if s[left] != s[right]:
                return False 
            else:
                left += 1 
                right -= 1 
        return True 
```

682 - Baseball Game

You're now a baseball game point recorder.

Given a list of strings, each string can be one of the 4 following types:

Integer (one round's score): Directly represents the number of points you get in this round.

"+" (one round's score): Represents that the points you get in this round are the sum of the last two valid round's points.

"D" (one round's score): Represents that the points you get in this round are the doubled data of the last valid round's points.

"C" (an operation, which isn't a round's score): Represents the last valid round's points you get were invalid and should be removed.

Each round's operation is permanent and could have an impact on the round before and the round after.

You need to return the sum of the points you could get in all the rounds.

Example 1:

Input: ["5","2","C","D","+"]

Output: 30

Explanation: 

Round 1: You could get 5 points. The sum is: 5.

Round 2: You could get 2 points. The sum is: 7.

Operation 1: The round 2's data was invalid. The sum is: 5.  

Round 3: You could get 10 points (the round 2's data has been removed). The sum is: 15.

Round 4: You could get 5 + 10 = 15 points. The sum is: 30.

```python
class Solution(object):
    def calPoints(self, ops):
        """
        :type ops: List[str]
        :rtype: int
        """
        if not ops:
            return 0 
        stack = []
        for string in ops:
            if string == 'C' and stack:
                stack.pop()
            elif string == 'D' and len(stack) >= 1:
                stack.append(2*stack[-1])
            elif string == '+' and len(stack) >= 2:
                stack.append(stack[-2] + stack[-1])
            else:
                stack.append(int(string))
        # print(stack)
        return sum(stack)
            
        
```

687 - Longest Univalue Path

Given a binary tree, find the length of the longest path where each node in the path has the same value. This path may or may not pass through the root.

The length of path between two nodes is represented by the number of edges between them.

 

Example 1:

Input:

              5
             / \
            4   5
           / \   \
          1   1   5
          
Output: 2

```python
# DFS
# use a helper function to find the max univalue path PASSING THROUGH the current node 
# use a global variable to record longest univalue path 
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def longestUnivaluePath(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        if not root:
            return 0 
        self.length = 0 
        self.helper(root)
        return self.length
    
    # find the longest consecutive length PASSING THROUGH node 
    def helper(self, node):
        if node is None:
            return 0
        left_len = self.helper(node.left)
        right_len = self.helper(node.right)
        # use left_max, right_max to record the longest univalue path on left and right subtree
        left_max, right_max = 0, 0 
        if node.left and node.val == node.left.val:
            left_max = left_len + 1 
        if node.right and node.val == node.right.val:
            right_max = right_len + 1 
        # if node.left and node.right and node.left.val == node.right.val == node.val:
        #     right_max -= 1
         # update longest univalue path    
        self.length = max(self.length, left_max + right_max)
        return max(left_max, right_max)
        
```

692 - Top K Frequent Words

Given a non-empty list of words, return the k most frequent elements.

Your answer should be sorted by frequency from highest to lowest. If two words have the same frequency, then the word with the lower alphabetical order comes first.

Example 1:
    
Input: ["i", "love", "leetcode", "i", "love", "coding"], k = 2
    
Output: ["i", "love"]

```python
# hashtable and heap 
# special case: ['a', 'aa', 'aaa'], k = 2, should have output ['a', 'aa']
from heapq import heappush, heappop, nsmallest 
class Solution(object):
    def topKFrequent(self, words, k):
        """
        :type words: List[str]
        :type k: int
        :rtype: List[str]
        """
        if not words:
            return []
        # sort the words so the indices can be used to label the order of words 
        words.sort(reverse = True)
        # count the frequency 
        hash_table = {}
        for word in words:
            hash_table.setdefault(word, 0)
            hash_table[word] += 1 
        
        # find top k frequent 
        heap = []
        for word in hash_table:
            # push word frequency, index (lower alphabetical words have bigger index, 
            # so it can be kept in the heap when two words have the same frequency and need 
            # to compare their index )
            heappush(heap, (hash_table[word], words.index(word), word))
            if len(heap) > k:
                heappop(heap)
       
        return [word for _, _, word in sorted(heap, key = lambda x: (-x[0], -x[1]))]
        
```
