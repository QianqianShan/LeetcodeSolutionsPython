617 - Merge Two Binary Trees

Given two binary trees and imagine that when you put one of them to cover the other, some nodes of the two trees are overlapped while the others are not.

You need to merge them into a new binary tree. The merge rule is that if two nodes overlap, then sum node values up as the new value of the merged node. Otherwise, the NOT null node will be used as the node of new tree.

Example 1:

Input: 
    
	Tree 1                     Tree 2                  
          1                         2                             
         / \                       / \                            
        3   2                     1   3                        
       /                           \   \                      
      5                             4   7                  

Output: 

Merged tree:

	     3
	    / \
	   4   5
	  / \   \ 
	 5   4   7

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

from collections import deque 
class Solution(object):
    def mergeTrees(self, t1, t2):
        """
        :type t1: TreeNode
        :type t2: TreeNode
        :rtype: TreeNode
        """
        if t1 is None:
            return t2
        if t2 is None:
            return t1
        t3 = TreeNode(t1.val + t2.val)
        t3.left = self.mergeTrees(t1.left, t2.left)
        t3.right = self.mergeTrees(t1.right, t2.right)
        return t3
        
        
```

621 - Task Scheduler

Given a char array representing tasks CPU need to do. It contains capital letters A to Z where different letters represent different tasks. Tasks could be done without original order. Each task could be done in one interval. For each interval, CPU could finish one task or just be idle.

However, there is a non-negative cooling interval n that means between two same tasks, there must be at least n intervals that CPU are doing different tasks or just be idle.

You need to return the least number of intervals the CPU will take to finish all the given tasks.

 

Example:

Input: tasks = ["A","A","A","B","B","B"], n = 2
    
Output: 8
    
Explanation: A -> B -> idle -> A -> B -> idle -> A -> B.

```python
# find the max freq of tasks
# the max len is (max_freq - 1) * (n + 1) + # of tasks that have this max_freq 
# n + 1: n intervals + task another time 
# AABBCD with n = 3: ABC_ABD  1 * 4 + 2
# if the above length < len(tasks), then the max interval is len(tasks)
class Solution(object):
    def leastInterval(self, tasks, n):
        """
        :type tasks: List[str]
        :type n: int
        :rtype: int
        """
        counts = {}
        for task in tasks:
            counts.setdefault(task, 0)
            counts[task] += 1 
        max_count_task = None 
        max_count = 0
        for task in counts:
            if counts[task] > max_count:
                max_count_task = task
                max_count = counts[task]
        num_max_count = len([k for k, v in counts.items() if v == max_count])
        # print(num_max_count)
        total = (max_count - 1) * (n + 1) + num_max_count 
        return max(total, len(tasks))
            
```

628 - Maximum Product of Three Numbers

Given an integer array, find three numbers whose product is maximum and output the maximum product.

Example 1:

Input: [1,2,3]
    
Output: 6
 

Example 2:

Input: [1,2,3,4]
    
Output: 24


```python
# sort the numbers
# no matter how many negative/positive values in nums
# the biggest three number product should be 
#  max(nums[0] * nums[1] * nums[-1], nums[-3] * nums[-2] * nums[-1])
class Solution(object):
    def maximumProduct(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        if not nums:
            return 
        nums.sort()
        if len(nums) == 3:
            return nums[0] * nums[1] * nums[2]
        n_positive = sum([i > 0 for i in nums])
        n_negative = len(nums) - n_positive 
        if n_positive == len(nums) or n_negative <= 1:
            return nums[-3] * nums[-2] * nums[-1]
        return max(nums[0] * nums[1] * nums[-1], nums[-3] * nums[-2] * nums[-1])
```

637 - Average of Levels in Binary Tree

Given a non-empty binary tree, return the average value of the nodes on each level in the form of an array.

Example 1:
    
Input:
    
      3
     / \
    9  20
    /  \
    15   7
    
Output: [3, 14.5, 11]

Explanation:
The average value of nodes on level 0 is 3,  on level 1 is 14.5, and on level 2 is 11. Hence return [3, 14.5, 11].

```python
# level order traversal with BFS 
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

from collections import deque 
class Solution(object):
    def averageOfLevels(self, root):
        """
        :type root: TreeNode
        :rtype: List[float]
        """
        if not root:
            return []
        # initialization 
        results = []
        queue = deque([root])
        # loop 
        while queue:
            size = len(queue)
            vals = []
            for _ in range(size):
                cur_node = queue.popleft()
                vals.append(cur_node.val)
                if cur_node.left:
                    queue.append(cur_node.left)
                if cur_node.right:
                    queue.append(cur_node.right)
            # add mean of each level to results
            # print(vals)
            results.append(sum(vals)/float(len(vals)))
        return results 
        
```

647 - Palindromic Substrings


Given a string, your task is to count how many palindromic substrings in this string.

The substrings with different start indexes or end indexes are counted as different substrings even they consist of same characters.

Example 1:

Input: "abc"

Output: 3

Explanation: Three palindromic strings: "a", "b", "c".
 

Example 2:

Input: "aaa"

Output: 6

Explanation: Six palindromic strings: "a", "a", "a", "aa", "aa", "aaa".

```python
# 1. dp, use dp[i][j] to record if s[i:j+1] is palindrome or not 
class Solution(object):
    def countSubstrings(self, s):
        """
        :type s: str
        :rtype: int
        """
        if not s:
            return 0 
        n = len(s)
        # initialization 
        dp = [[0 for _ in range(n)] for _ in range(n)]
        # for i in range(n):
        #     dp[i][i] = 1 
        # the total number of palindrome substrings 
        total = 0
        for i in range(n):
            for j in range(i + 1):
                if s[i] == s[j] and (i - j <= 2 or dp[j + 1][i - 1] == 1):
                    dp[j][i] = 1 
                total += dp[j][i]
        return total 
                
```

```python
# 2. dp, use dp[i][j] to record if s[i:j+1] is palindrome or not 
class Solution(object):
    def countSubstrings(self, s):
        """
        :type s: str
        :rtype: int
        """
        if not s:
            return 0 
        n = len(s)
        # initialization 
        dp = [[0 for _ in range(n)] for _ in range(n)]
        for i in range(n):
            dp[i][i] = 1 
            dp[i][i - 1] = 1 
            
        # the total number of palindrome substrings 
        # each single character is palindrome 
        total = n
        # add palindrome wight length > 1 
        for length in range(1, n + 1):
            for start in range(n - length):
                end = start + length
                if s[start] == s[end] and dp[start + 1][end - 1]:
                    dp[start][end] = 1 
                    total += dp[start][end]
        return total 
                
```
