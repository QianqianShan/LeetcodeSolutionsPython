1010 - Pairs of Songs With Total Durations Divisible by 60

In a list of songs, the i-th song has a duration of time[i] seconds. 

Return the number of pairs of songs for which their total duration in seconds is divisible by 60.  Formally, we want the number of indices i < j with (time[i] + time[j]) % 60 == 0.

 

Example 1:

Input: [30,20,150,100,40]
    
Output: 3
    
Explanation: Three pairs have a total duration divisible by 60:
        
(time[0] = 30, time[2] = 150): total duration 180
    
(time[1] = 20, time[3] = 100): total duration 120
    
(time[1] = 20, time[4] = 40): total duration 60

```python
# 1. Use hash table to frequencies of numbers of their residuals after divided by 60 
# 2. count each possible pair
class Solution(object):
    def numPairsDivisibleBy60(self, time):
        """
        :type time: List[int]
        :rtype: int
        """
        if not time:
            return 0 
        hash_table = {}
        for item in time:
            element = item % 60 
            if element not in hash_table:
                hash_table[element] = 1
            else:
                hash_table[element] += 1 
        # print(hash_table)
        count = 0 
        if 0 in hash_table:
            count += (hash_table[0] * (hash_table[0] - 1))//2
        if 30 in hash_table:
            count += (hash_table[30] * (hash_table[30] - 1))//2
        for i in range(1, 30):
            if i in hash_table and (60 - i) in hash_table:
                # print(i)
                count += hash_table[i] * hash_table[60 - i]
        return count 
        
```

1022 - Sum of Root To Leaf Binary Numbers

Given a binary tree, each node has value 0 or 1.  Each root-to-leaf path represents a binary number starting with the most significant bit.  For example, if the path is 0 -> 1 -> 1 -> 0 -> 1, then this could represent 01101 in binary, which is 13.

For all leaves in the tree, consider the numbers represented by the path from the root to that leaf.

Return the sum of these numbers.



```python
# dfs 
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def sumRootToLeaf(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        if not root:
            return 0 
        results = []
        self.dfs(root, str(root.val), results)
        # print(results)
        return sum([int(''.join(path), base = 2) for path in results])
    def dfs(self, node, path, results):
        if node is None:
            return 
        if node.left is None and node.right is None:
            results.append(path)
            return 
        if node.left is not None:
            self.dfs(node.left, path + str(node.left.val), results)
        if node.right is not None:
            self.dfs(node.right, path + str(node.right.val), results)
        
        
```

1029 - Two City Scheduling

There are 2N people a company is planning to interview. The cost of flying the i-th person to city A is costs[i][0], and the cost of flying the i-th person to city B is costs[i][1].

Return the minimum cost to fly every person to a city such that exactly N people arrive in each city.

 

Example 1:

Input: [[10,20],[30,200],[400,50],[30,20]]
    
Output: 110

```python
# Greedy, sort the data by the costs differences of each person to A and to B, 
class Solution(object):
    def twoCitySchedCost(self, costs):
        """
        :type costs: List[List[int]]
        :rtype: int
        """
        if not costs:
            return 
        # sort costs by the differences between cost to A and cost to B 
        costs.sort(key = lambda x: x[0] - x[1])
        # send the first half to A and the last half to B 
        N = len(costs) // 2 
        total = 0 
        for i in range(N):
            total += costs[i][0]
            # print(total)
        for i in range(N, 2*N):
            total += costs[i][1]
        return total 
```

1046 - Last Stone Weight

We have a collection of rocks, each rock has a positive integer weight.

Each turn, we choose the two heaviest rocks and smash them together.  Suppose the stones have weights x and y with x <= y.  The result of this smash is:

If x == y, both stones are totally destroyed;

If x != y, the stone of weight x is totally destroyed, and the stone of weight y has new weight y-x.

At the end, there is at most 1 stone left.  Return the weight of this stone (or 0 if there are no stones left.)

 

Example 1:

Input: [2,7,4,1,8,1]
    
Output: 1

```python
# use heap to avoid consistently sorting 
from heapq import heappush, heappop 
class Solution(object):
    def lastStoneWeight(self, stones):
        """
        :type stones: List[int]
        :rtype: int
        """
        if not stones:
            return 0 
        heap = []
        # add elements to maxheap 
        for stone in stones:
            heappush(heap, -stone)
        while len(heap) > 1:
            w1 = -heappop(heap)
            w2 = -heappop(heap)
            diff = (w1 - w2) if w1 > w2 else (w2 - w1)
            if diff > 0:
                heappush(heap, -diff)
        if len(heap) == 1:
            return -heap[0]
        else:
            return 0
            
        
```

1047 -  Remove All Adjacent Duplicates In String

Given a string S of lowercase letters, a duplicate removal consists of choosing two adjacent and equal letters, and removing them.

We repeatedly make duplicate removals on S until we no longer can.

Return the final string after all such duplicate removals have been made.  It is guaranteed the answer is unique.

 

Example 1:

Input: "abbaca"
    
Output: "ca"

```python
class Solution(object):
    def removeDuplicates(self, S):
        """
        :type S: str
        :rtype: str
        """
        if not S:
            return ''
        stack = []
        i = 0 
        while i < len(S):
            # print(stack)
            # remove duplicates 
            while stack and i < len(S) and stack[-1] == S[i]:
                stack.pop()
                i += 1 
            # add S[i]
            if i < len(S):
                stack.append(S[i])
                i += 1 
        return ''.join(stack)
        
```

1050 - Actors and Directors Who Cooperated At Least Three Times

Write a SQL query for a report that provides the pairs (actor_id, director_id) where the actor have cooperated with the director at least 3 times.

```python
# Write your MySQL query statement below
# add new column as number of times of cooperation 
# filter 

select actor_id, director_id 
from (
    select actor_id, director_id, count(*)  as count 
    from ActorDirector 
    group by 1, 2
    order by 1, 2 ) tmp 
where count >= 3 
```
